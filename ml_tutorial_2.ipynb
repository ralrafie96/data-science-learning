```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("./tutorial_2/homeprices.csv")
df
```

```python
%matplotlib inline
plt.xlabel('area (sqft)')
plt.ylabel('price (USD)')
plt.scatter(df.area, df.price, color="red", marker="+")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[["area"]], df["price"])
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
y = reg.coef_ * 3300 + reg.intercept_
y
```

```python
%matplotlib inline
plt.xlabel('area (sqft)')
plt.ylabel('price (USD)')
plt.scatter(df.area, df.price, color="red", marker="+")
plt.plot(df.area, reg.predict(df[['area']]), color="blue")
```

```python
d = pd.read_csv('tutorial_2/areas.csv')
d
```

```python
p = reg.predict(d)
d['price'] = p
d
```

```python
d.to_csv('tutorial_2/areas2.csv', index=False)
```
